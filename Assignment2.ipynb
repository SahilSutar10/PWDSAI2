```python
'''1] Discuss string slicing and provide examples.
Slicing means from where we have to see the output.
In other words, String slicing in Python is a manner to extract specific components of a string. It allows you to pick out substrings using a number of indices, making it useful for manipulating text.
The general syntax for string slicing is :
string[start:end:step]
a] start: The index at which the slice starts (inclusive).
b] end: The index at which the slice ends (exclusive).
c] step: The interval between each index in the slice (optional, default is 1).
'''
#Example 1 : Basic Example
name = "Sahil Sutar"
print(name[0:5])  # because H is at '0th'place and O is at '4th'place and 5 is exclusive. That's why, Output: 'Sahil'

#Example 2: Omitting Start or End
print(name[:5])    # Output: 'Sahil' (from start up to index 5)
print(name[6:])    # Output: 'Sutar' (from index 6 to end)

#Example 3: Extracting Substrings with Negative Steps
print(name[-1:-6:-1])  # Output: 'ratuS' (last 5 characters in reverse)

#Example 4: Reversing a String
print(name[::-1])  # Output: 'ratuS lihaS'

#Example 5: Using the Step Argument
print(name[0:10:2])  # Output: 'ShlSt' (every 2nd character from index 0 to 10)


```

```python
'''2] Explain the key features of lists in Python.
i)Lists in Python are one of the maximum flexible and broadly used data structures.
ii) They let you store collections of gadgets in a single variable and provide good sized capability for data manipulation.
Here are some key features of lists in python :
a) It is ordered collection of items.
b) It is mutuable. Meaning you can change, add, or remove items after the list has been created.
c) Lists can include duplicate elements, this means that the identical value can seem greater than once.
d) Lists are heterogeneous. It can holds elements of different data types, such as strings, integers, floats, or even other lists.
e) Lists support indexing (both positive and negative) and slicing, making it easy to access and manipulate subsets of elements.
f) Lists in Python are dynamic, on the way to develop or decrease as wanted via including or casting off elements.
g) Comprehensive Methods :
Lists come with numerous built-in methods for manipulating the list. Some commonly used ones include:
append(): Adds an element to the end.
insert(): Adds an element at a specified position.
remove(): Removes the first occurrence of a specified element.
pop(): Removes and returns an element at a specified position (default is the last item).
sort(): Sorts the list in ascending or descending order.
reverse(): Reverses the list order.
count(): Counts the occurrences of an element.
index(): Finds the index of the first occurrence of an element.'''
# Creating a list of names
names = ["Sahil", "Pooja", "Priya"]
print("Original list:", names)  # Output: ['Sahil', 'Pooja', 'Priya']

# Adding a new name to the list (Mutability)
names.append("Amit")
print("After appending 'Amit':", names)  # Output: ['Sahil', 'Pooja', 'Priya', 'Amit']

# Accessing items using index
print("First name:", names[0])  # Output: 'Sahil'
print("Second name:", names[1])  # Output: 'Pooja'

# Slicing the list
print("Sliced list (first two names):", names[:2])  # Output: ['Sahil', 'Pooja']

# Modifying an item in the list (Mutability)
names[2] = "Priyanka"
print("After modifying 'Priya' to 'Priyanka':", names)  # Output: ['Sahil', 'Pooja', 'Priyanka', 'Amit']

# Using list comprehension to create a new list with all names in lowercase
lowercase_names = [name.lower() for name in names]
print("Lowercase names:", lowercase_names)  # Output: ['sahil', 'pooja', 'priyanka', 'amit']

# Removing an item from the list
names.remove("Amit")
print("After removing 'Amit':", names)  # Output: ['Sahil', 'Pooja', 'Priyanka']

# Checking the list size (Dynamic nature)
print("Number of names:", len(names))  # Output: 3

# Nested list example (List of lists)
nested_names = [["Sahil", "Pooja"], ["Priya", "Amit"]]
print("Nested list:", nested_names)  # Output: [['Sahil', 'Pooja'], ['Priya', 'Amit']]
print("Accessing nested list item:", nested_names[1][0])  # Output: 'Priya'
```

```python
'''3]  Describe how to access, modify, and delete elements in a list with examples.'''
#1. Accessing Elements in a List
# Example list
names = ["Sahil", "Pooja", "Priya", "Amit"]

# Accessing a single element using index
print(names[0])  # Output: 'Sahil' (first element)
print(names[2])  # Output: 'Priya' (third element)

# Accessing the last element using negative index
print(names[-1])  # Output: 'Amit'

# Accessing a range of elements using slicing
print(names[1:3])  # Output: ['Pooja', 'Priya'] (from index 1 to 2)

#2. Modifying Elements in a List
#Since lists are mutable, you can change individual elements by assigning a new value to a specific index.
# Modifying a single element
names[1] = "Neha"
print(names)  # Output: ['Sahil', 'Neha', 'Priya', 'Amit']

# Modifying multiple elements using slicing
names[1:3] = ["Pooja", "Priyanka"]
print(names)  # Output: ['Sahil', 'Pooja', 'Priyanka', 'Amit']

'''Deleting Elements in a List
There are several ways to delete elements in a list:

Using del statement: Deletes elements at a specific index or slice.
Using remove() method: Deletes the first occurrence of a specific value.
Using pop() method: Removes and returns an element at a specific index (defaults to the last element if no index is provided).'''
#Using del Statement
# Using del to delete an element by index
del names[1]
print(names)  # Output: ['Sahil', 'Priyanka', 'Amit']

# Using del to delete a slice of elements
names = ["Sahil", "Pooja", "Priya", "Amit"]
del names[1:3]
print(names)  # Output: ['Sahil', 'Amit']

#Using remove() Method
# Using remove to delete an element by value
names = ["Sahil", "Pooja", "Priya", "Amit"]
names.remove("Priya")
print(names)  # Output: ['Sahil', 'Pooja', 'Amit']

#Using pop() Method
# Using pop to remove the last element
last_element = names.pop()
print(last_element)  # Output: 'Amit'
print(names)         # Output: ['Sahil', 'Pooja']

# Using pop to remove an element at a specific index
second_element = names.pop(1)
print(second_element)  # Output: 'Pooja'
print(names)           # Output: ['Sahil']

```

```python
'''4] Compare and contrast tuples and lists with examples.
Tuples and lists are both data structures in Python that store ordered collections of items.
However, they have some significant differences in terms of mutability, syntax, and usage.
Here’s a comparison of tuples and lists:
a) Mutability
i) Lists are mutable, meaning you can modify, add, or remove elements after the list is created.
ii) Tuples are immutable, meaning once a tuple is created, you cannot modify its elements, add, or remove items.'''
# List example (mutable)
names_list = ["Sahil", "Pooja", "Priya"]
names_list[1] = "Neha"  # Modifying an element
names_list.append("Amit")  # Adding a new element
print(names_list)  # Output: ['Sahil', 'Neha', 'Priya', 'Amit']

# Tuple example (immutable)
names_tuple = ("Sahil", "Pooja", "Priya")
# names_tuple[1] = "Neha"  # This will raise a TypeError because tuples are immutable

'''b) Syntax
i) Lists are defined using square brackets [].
ii) Tuples are defined using parentheses ().'''
# List
my_list = [1, 2, 3]

# Tuple
my_tuple = (1, 2, 3)

'''c) Performance
i) Tuples are generally more memory-efficient and faster to access than lists because they are immutable.
ii) Lists consume more memory and are slower due to their dynamic nature (allowing modifications).

d) Usage
i) Lists are used when you need a collection of items that can change over time (e.g., adding or removing elements).
ii) Tuples are used for fixed collections of items that shouldn't change, like coordinates, fixed sets of options, or data that’s constant.'''
# List (used when elements may change)
shopping_list = ["milk", "eggs", "bread"]
shopping_list.append("butter")
print(shopping_list)  # Output: ['milk', 'eggs', 'bread', 'butter']

# Tuple (used for fixed data)
coordinates = (40.7128, -74.0060)  # Example of geographic coordinates
print(coordinates)  # Output: (40.7128, -74.0060)

'''e) Functions and Methods
i) Lists have numerous methods for manipulation (append(), remove(), sort(), pop(), etc.).
ii) Tuples have only a few methods, mainly for counting and finding elements (count() and index()), as they can’t be modified.'''
# List methods
numbers_list = [10, 20, 30]
numbers_list.append(40)
numbers_list.remove(20)
print(numbers_list)  # Output: [10, 30, 40]

# Tuple methods
numbers_tuple = (10, 20, 30, 10)
print(numbers_tuple.count(10))  # Output: 2 (count occurrences of 10)
print(numbers_tuple.index(20))  # Output: 1 (find index of 20)

```

```python
'''5] Describe the key features of sets and provide examples of their use.
Sets in Python are a type of collection similar to lists and tuples but with distinct features that make them ideal for certain operations, especially those involving uniqueness and mathematical set operations.
 Sets are defined by {} this including in it some values or anything otherwise it willl become dictionary.
 Here are the key features of sets, along with examples:
a) Unordered Collection
i) Sets are unordered, meaning the items do not have a defined order, and they do not support indexing or slicing.
ii) This is unlike lists and tuples, which maintain the order of elements.

b) No Duplicate Elements
i) Sets automatically remove duplicates, so each element in a set is unique.
ii) This is helpful when you need a collection of unique items, like eliminating duplicates from a list.

c)Mutable
i)Sets are mutable, meaning you can add or remove items after the set is created.
ii) However, only immutable (hashable) objects, like numbers, strings, and tuples, can be added as set elements (no lists or dictionaries).

d) Optimized for Membership Tests
i) Sets are highly optimized for checking if an element exists in the set.
ii) Membership tests (using in or not in) are faster in sets compared to lists.

e) Mathematical Set Operations
i) Sets support common mathematical set operations like union, intersection, difference, and symmetric difference.
ii) These operations are useful for tasks that involve comparisons between multiple sets.'''

#Examples of Set Operations
#1] Adding and Removing Elements
names = {"Sahil", "Pooja", "Priya"}
names.add("Amit")  # Adding an element
print(names)  # Output: {'Sahil', 'Pooja', 'Priya', 'Amit'}

names.remove("Pooja")  # Removing an element
print(names)  # Output: {'Sahil', 'Priya', 'Amit'}

#2] Membership Test
print("Sahil" in names)  # Output: True
print("Pooja" in names)   # Output: False

#3] Mathematical Set Operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union (all elements from both sets)
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection (common elements)
print(set_a & set_b)  # Output: {3, 4}

# Difference (elements in set_a but not in set_b)
print(set_a - set_b)  # Output: {1, 2}

# Symmetric Difference (elements in either set_a or set_b but not in both)
print(set_a ^ set_b)  # Output: {1, 2, 5, 6}

#4] Finding Common and Unique Elements in Multiple Collections
students_in_math = {"Sahil", "Pooja", "Priya"}
students_in_science = {"Priya", "Amit", "Neha"}

# Students enrolled in both math and science
both_courses = students_in_math & students_in_science
print(both_courses)  # Output: {'Priya'}

# Students only in math
only_math = students_in_math - students_in_science
print(only_math)  # Output: {'Sahil', 'Pooja'}

#5] Removing Duplicates from a List
numbers_list = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers_list)
print(unique_numbers)  # Output: {1, 2, 3, 4, 5}







```

```python
'''7]  Describe how to add, modify, and delete items in a dictionary with examples.
Dictionaries in Python are collections of key-value pairs that allow you to store and manipulate data by using keys.Keys are unique and immmutable.
 Here’s how to add, modify, and delete items in a dictionary, with examples.
 a) Adding Items to a Dictionary
To add a new key-value pair to a dictionary, you can use the following methods:
Direct assignment: Assign a value to a new key.
update() method: Adds multiple key-value pairs at once.'''
# Initial dictionary
student = {"name": "Sahil", "age": 20}

# Adding a new key-value pair directly
student["major"] = "Computer Science"
print(student)
# Output: {'name': 'Sahil', 'age': 20, 'major': 'Computer Science'}

# Using update() to add multiple key-value pairs
student.update({"grade": "A", "year": 2})
print(student)
# Output: {'name': 'Sahil', 'age': 20, 'major': 'Computer Science', 'grade': 'A', 'year': 2}

#b) Modifying Items in a Dictionary
#To modify an existing key’s value, assign a new value to the key.
#If the key exists, its value is updated; if it doesn’t, it’s added as a new key.
# Initial dictionary
student = {"name": "Sahil", "age": 20, "major": "Computer Science"}

# Modifying the value of an existing key
student["age"] = 21
print(student)
# Output: {'name': 'Sahil', 'age': 21, 'major': 'Computer Science'}

# Modifying multiple values using update()
student.update({"major": "Data Science", "grade": "A"})
print(student)
# Output: {'name': 'Sahil', 'age': 21, 'major': 'Data Science', 'grade': 'A'}


'''c) Deleting Items from a Dictionary
There are several ways to delete items from a dictionary:

del statement: Deletes a specific key or the entire dictionary.
pop() method: Removes a key and returns its value.
popitem() method: Removes and returns the last inserted key-value pair.
clear() method: Empties the dictionary, leaving it empty.'''
# Initial dictionary
student = {"name": "Sahil", "age": 21, "major": "Data Science", "grade": "A"}

# Using del to delete a specific key
del student["grade"]
print(student)
# Output: {'name': 'Sahil', 'age': 21, 'major': 'Data Science'}

# Using pop() to remove a key and get its value
age = student.pop("age")
print(age)     # Output: 21
print(student) # Output: {'name': 'Sahil', 'major': 'Data Science'}

# Using popitem() to remove the last inserted key-value pair
last_item = student.popitem()
print(last_item)  # Output: ('major', 'Data Science')
print(student)    # Output: {'name': 'Sahil'}

# Using clear() to empty the dictionary
student.clear()
print(student)    # Output: {}

```

```python
'''8] Discuss the importance of dictionary keys being immutable and provide examples.
In Python, dictionary keys ought to be immutable due to the fact they want to be hashable.
 The immutability of dictionary keys ensures that their values do not alternate even as they're being used as keys, that's vital for the dictionary's internal structure and overall performance
Let’s break down why dictionary keys need to be immutable and provide examples.
a) Hashing Mechanism:
In Python, dictionaries use a hash table internally to store key-value pairs. The hash of a key is computed when the key is inserted into the dictionary.
Immutable objects (like strings, numbers, and tuples) have a fixed hash value, which doesn’t change over time.
This allows the dictionary to reliably locate the key when performing operations like lookups, insertions, and deletions.

b) Consistency:
If dictionary keys have been mutable (e.g., lists or dictionaries), their hash value could alternate after being inserted into the dictionary.
This would motive troubles due to the fact the dictionary’s internal structure might not be capable of locate the important thing (seeing that its hash has changed), leading to unexpected conduct and mistakes.

c) Performance:
The use of hashable (immutable) keys allows dictionaries to perform lookups, insertions, and deletions efficiently in O(1) time on average, which is one of the key advantages of using a dictionary over other data structures.

You would typically use immutable keys when you need to create a dictionary where each key represents a unique, constant entity that doesn’t change, such as:
Identifiers: A unique string or number as the key (e.g., usernames, product IDs).
Geographical coordinates: Using tuples (latitude, longitude) as keys.
Fixed attributes: Using tuples to represent combinations of values that should remain unchanged, like RGB values for colors.'''

#Examples of Immutable Keys
#Immutable Keys
#Strings, numbers, and tuples are immutable, meaning they can be used as dictionary keys:
# Strings as dictionary keys
student = {"name": "Sahil", "age": 21}
print(student["name"])  # Output: Sahil

# Numbers as dictionary keys
price = {100: "Apple", 200: "Banana", 300: "Cherry"}
print(price[200])  # Output: Banana

# Tuples as dictionary keys (since they are immutable)
coordinates = {(40.7128, -74.0060): "New York", (34.0522, -118.2437): "Los Angeles"}
print(coordinates[(40.7128, -74.0060)])  # Output: New York


```
